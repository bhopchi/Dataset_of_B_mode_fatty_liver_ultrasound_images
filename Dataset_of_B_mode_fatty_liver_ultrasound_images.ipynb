```python
import scipy.io
fname='/content/dataset_liver_bmodes_steatosis_assessment_IJCARS.mat'
mat = scipy.io.loadmat(fname)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
mat.keys()
```

```python
# Extract the 'data' from the mat file
data = mat['data']

# Check the shape of the data
print(data.shape)
```

```python
for i in data:
    print(i)
```

```python
first_element = data[0]
# Check the type of this element
print(type(first_element))
```

```python
print(len(first_element))
```

```python
# Explore the contents of the first element
print(first_element[0][3].shape)
```

```python
plt.imshow(first_element[0][3][1])
```

```python
first_element[0][3].shape[0]
```

```python
for i in range(10):
    plt.title(f"Label:{first_element[i][0]} Size:{first_element[i][3][0].shape}")
    plt.imshow(first_element[i][3][0], cmap = "gray")
    plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=5, figsize=(15, 10))
for j in range(5):
    for i in range(2):
        axs[i,j].set_title(f"Label:{first_element[i+j][0]}")
        axs[i,j].imshow(first_element[i+j][3][0], cmap = "gray")

plt.tight_layout()
plt.show()
```
